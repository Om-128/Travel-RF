## Holiday Package Prediciton

### 1) Problem statement.
"Trips & Travel.Com" company wants to enable and establish a viable business model to expand the customer base.
One of the ways to expand the customer base is to introduce a new offering of packages. Currently, there are 5 types of packages the company is offering * Basic, Standard, Deluxe, Super Deluxe, King. Looking at the data of the last year, we observed that 18% of the customers purchased the packages. However, the marketing cost was quite high because customers were contacted at random without looking at the available information.
The company is now planning to launch a new product i.e. Wellness Tourism Package. Wellness Tourism is defined as Travel that allows the traveler to maintain, enhance or kick-start a healthy lifestyle, and support or increase one's sense of well-being.
However, this time company wants to harness the available data of existing and potential customers to make the marketing expenditure more efficient.
### 2) Data Collection.
The Dataset is collected from https://www.kaggle.com/datasets/susant4learning/holiday-package-purchase-prediction
The data consists of 20 column and 4888 rows.

```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
import plotly.express as px
import warnings

warnings.filterwarnings("ignore")

%matplotlib inline
```

```python
df = pd.read_csv('Travel.csv')
df.head()
```

# Data clearning
    
1. Handling missing values
2. Handling duplicates
3. Check data types
4. Understand the dataset

```python
df.isnull().sum()
```

```python
### Check all the categories 
df['Gender'].value_counts()
```

```python
# Handle gender
df['Gender'] = df['Gender'].replace('Fe Male', 'Female')
df['MaritalStatus'] = df['MaritalStatus'].replace('Single', 'Unamrried')
```

```python
df.head()
```

```python
#Checking Missing values

features_with_na = [feature for feature in df.columns if df[feature].isnull().sum() >= 1]
for feature in features_with_na:
    print(feature, np.round(df[feature].isnull().mean()*100,5), '% missing values')
```

# Replacing null values

```python
# Age
df.Age.fillna(df.Age.median(), inplace=True)

#Type of contract
df.TypeofContact.fillna(df.TypeofContact.mode()[0], inplace = True)

#Duration of pitch
df.DurationOfPitch.fillna(df.DurationOfPitch.median(), inplace=True)

#NumberOfFollowUps
df.NumberOfFollowups.fillna(df.NumberOfFollowups.mode()[0], inplace=True)
```

```python
#Prefered Property
df.PreferredPropertyStar.fillna(df.PreferredPropertyStar.mode()[0], inplace=True)

#Number of trips
df.NumberOfTrips.fillna(df.NumberOfTrips.median(), inplace=True)

#Number of children visiting
df.NumberOfChildrenVisiting.fillna(df.NumberOfChildrenVisiting.mode()[0], inplace=True)

#Monthly Income
df.MonthlyIncome.fillna(df.MonthlyIncome.median(), inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop('CustomerID', inplace = True, axis=1)
```

# Feature Engineering

```python
#New column for visitors feature 

df['TotalVisitors'] = df['NumberOfPersonVisiting'] + df['NumberOfChildrenVisiting']
df.drop(columns=['NumberOfPersonVisiting', 'NumberOfChildrenVisiting'], inplace=True, axis=1)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
x = df.drop(['ProdTaken'], axis=1)
y = df['ProdTaken']
```

```python
y.value_counts()
```

```python
# Split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
x_train.shape, x_test.shape
```

```python
#Create column transformer with three types of transformer

cat_feature = x.select_dtypes(include="object").columns
num_features = x.select_dtypes(exclude="object").columns
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
numeric_transformer = StandardScaler()
OHE_transformer = OneHotEncoder(drop='first')

preprocessor = ColumnTransformer(
    [
        ("OneHotEncoder", OHE_transformer, cat_feature),
        ("StandardScalar", numeric_transformer, num_features)
    ]
)
```

```python
# Apply transformation on training dataset(use fit_transform())
x_train = preprocessor.fit_transform(x_train)
```

```python
# Apply transformation on test dataset(use transform())
x_test = preprocessor.transform(x_test)
```

# Training Model (Random Forest Classifier)

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay, \
                            precision_score, recall_score, f1_score, roc_curve
```

```python
models = {
    "Decision Tree" : DecisionTreeClassifier(),
    "Random Forest" : RandomForestClassifier()
}

#Train model

for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(x_train, y_train)

# Make Predictions

y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)
```

```python
#Calculate score

# Training set performance

model_train_accuracy = accuracy_score(y_train, y_train_pred)
mode_train_f1 = f1_score(y_train, y_train_pred)
mode_train_precision = precision_score(y_train, y_train_pred)
model_train_recall = recall_score(y_train, y_train_pred)

print("Training Set Perfomance")
print(model_train_accuracy)
print("_________________________")
print(mode_train_f1)
print("_________________________")
print(mode_train_precision)
print("_________________________")
print(model_train_recall)
```

```python
model_test_accuracy = accuracy_score(y_test, y_test_pred)
mode_test_f1 = f1_score(y_test, y_test_pred)
mode_test_precision = precision_score(y_test, y_test_pred)
model_test_recall = recall_score(y_test, y_test_pred)

print("Testing Set Perfomance")
print(model_test_accuracy)
print("_________________________")
print(mode_test_f1)
print("_________________________")
print(mode_test_precision)
print("_________________________")
print(model_test_recall)
```
